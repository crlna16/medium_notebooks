```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

import numpy as np
```

```python
sns.set_style('whitegrid')
```

Download the dataset here:

https://www.kaggle.com/datasets/jjayfabor/lettuce-growth-days?resource=download

```python
df = pd.read_csv('./data/lettuce_dataset.csv')
```

```python
df
```

```python
sns.lineplot(data=df, x='Growth Days', y='Humidity (%)', hue='Plant_ID', alpha=0.1, palette='Greens')

plt.savefig('./plots/2023-10-06_lettuce_humidity_all.png', dpi=150, bbox_inches='tight')
```

```python
sns.lineplot(data=df, x='Growth Days', y='Humidity (%)')

plt.savefig('./plots/2023-10-06_lettuce_humidity_mean.png', dpi=150, bbox_inches='tight')
```

```python
sns.lineplot(data=df, x='Growth Days', y='Humidity (%)')

plt.savefig('./plots/2023-10-06_lettuce_humidity_mean.png', dpi=150, bbox_inches='tight')
```

```python
sns.lineplot(data=df, x='Growth Days', y='Humidity (%)', errorbar=("ci", 75), label='CI 75%')
sns.lineplot(data=df, x='Growth Days', y='Humidity (%)', errorbar=("ci", 90), label='CI 90%')
sns.lineplot(data=df, x='Growth Days', y='Humidity (%)', errorbar=("ci", 95), label='CI 95%')

plt.savefig('./plots/2023-10-06_lettuce_humidity_mean_allci.png', dpi=150, bbox_inches='tight')
```

```python
sns.lineplot(data=df, x='Growth Days', y='Humidity (%)', marker='o', errorbar=("ci", 99), err_style="bars")

# â€œciâ€, â€œpiâ€, â€œseâ€, or â€œsdâ€

plt.savefig('./plots/2023-10-06_lettuce_humidity_mean_box.png', dpi=150, bbox_inches='tight')
```
